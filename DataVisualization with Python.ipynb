Features
    
- timestamp - timestamp field for grouping the data
- cnt - the count of a new bike shares
- t1 - real temperature in C
- t2 - temperature in C “feels like”
- hum - humidity in percentage
- wind_speed - wind speed in km/h

"weather_code" category description:

- 1 = Clear ; mostly clear but have some values with haze/fog/patches of fog/ fog in vicinity
- 2 = scattered clouds / few clouds

- 4 = Cloudy
- 7 = Rain/ light Rain shower/ Light rain
- 10 = rain with thunderstorm
- 26 = snowfall

<span style=" color:black"> Bike Sharing Data Analysis
    
* The bike sharing data contains information about the number of bike rentals per hour, along with various attributes such as weather, season, temperature, humidity, and wind speed. The data has no missing values and covers a period of three years (2015, 2016 and 2017). The main objective of this analysis is to explore the factors that affect the demand for bike rentals and to identify the patterns and trends in the data.

The following are some of the key insights derived from the data:

- The demand for bike rentals is higher on weekdays and working days than on weekends and holidays. This suggests that most of the bike users are commuters who use bikes as a mode of transportation to and from work or school.
- The season variable shows a moderate correlation with the month and temperature variables. This indicates that the seasonality of bike rentals is influenced by the changes in weather and temperature throughout the year.
- The summer season has the highest demand for bike rentals, followed by fall, spring, and winter. This implies that bike users prefer warmer and sunnier days for riding bikes. The winter season has the lowest demand, possibly due to the cold and snowy weather conditions that make biking less comfortable and safe.
- The peak hours for bike rentals are 8 am and 5 pm, which correspond to the typical working hours of most people. This confirms that bike users are mainly commuters who use bikes during rush hours to avoid traffic congestion and save time and money.
- The weather variable indicates the type of weather condition on a given day, such as clear, cloudy, light rain, or heavy rain. The data shows that bike users prefer clear or scattered clouds weather conditions for biking, as they have the highest demand for bike rentals. On the other hand, bike users avoid biking in rainy or stormy weather conditions, as they have the lowest demand for bike rentals.
